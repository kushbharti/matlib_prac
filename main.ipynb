```python
import matplotlib.pyplot as plt
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
sales = [10000, 5214, 6325, 8456, 9546, 4587, 5631, 4562, 9546, 10256, 5214, 8456]

plt.plot(months, sales, color='green', linestyle='--', linewidth=2, marker='o', label='2025 Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales per month')
plt.title('Monthly Sales Data Report', color='blue')
plt.legend(loc='lower right', fontsize=8,)
plt.grid(color='grey', linewidth=2)
plt.xlim(0, 11.5)
plt.ylim(4000,11000)
plt.show()
```

```python
# Bar Chat:-

products = ['A', 'B', 'C', 'D', 'E']
sales = [200, 175, 189, 112, 89]


plt.bar(products, sales, label='Sales 2025', color=['orange', 'blue', 'green', 'yellow', 'red'])
plt.xlabel('Products')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# Pie Chart:-

regions = ['North', 'South', 'East', 'West']
revenue = [3000, 2000, 1500, 1000]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', colors=['red', 'orange', 'gold', 'coral'])
plt.title('Revenue Contribution by Region')
plt.legend()
plt.axis('equal') 
plt.show()
```

```python
# Histogram :- 

scores = [85,95,62,41,53,15,47,51,26,48,75,95,42,63,35,15,85]

plt.hist(scores, bins=5, color='skyblue', edgecolor='black')
plt.xlabel('Score Range')
plt.ylabel('No. of Students')
plt.title('Score Distribution of Students')
plt.show()

```

```python
# Scatter Plot :- 

hours_studies = [1, 2, 3, 4, 5, 6, 7, 8]
exam_scores = [50, 55, 60, 65, 70, 75, 80, 85]

plt.scatter(hours_studies, exam_scores, color='blue', marker='^', label='Student Data')
plt.xlabel('Hours')
plt.ylabel('Exam Score')
plt.title('Relationship Between Study Time and Exam Score')
plt.legend()
plt.grid()
plt.show()
```

```python

plt.scatter([1, 2, 3], [50, 55, 60], color='blue', label='Group A')
plt.scatter([1, 2, 3], [45, 50, 52], color='orange', label='Group B')


plt.xlabel('Hours')
plt.ylabel('Exam Score')
plt.title('Comparison of Two Classes')
plt.legend()
plt.grid()
plt.show()
```
