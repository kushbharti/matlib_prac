```python
import matplotlib.pyplot as plt
```

```python
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sept', 'Oct', 'Nov', 'Dec']
sales = [10000, 5214, 6325, 8456, 9546, 4587, 5631, 4562, 9546, 10256, 5214, 8456]

plt.plot(months, sales, color='green', linestyle='--', linewidth=2, marker='o', label='2025 Sales Data')
plt.xlabel('Months')
plt.ylabel('Sales per month')
plt.title('Monthly Sales Data Report', color='blue')
plt.legend(loc='lower right', fontsize=8,)
plt.grid(color='grey', linewidth=2)
plt.xlim(0, 11.5)
plt.ylim(4000,11000)
plt.show()
```

```python
# Bar Chat:-

products = ['A', 'B', 'C', 'D', 'E']
sales = [200, 175, 189, 112, 89]


plt.bar(products, sales, label='Sales 2025', color=['orange', 'blue', 'green', 'yellow', 'red'])
plt.xlabel('Products')
plt.ylabel('Sales')
plt.legend()
plt.show()
```

```python
# Pie Chart:-


regions = ['North', 'South', 'East', 'West']
revenue = [3000, 2000, 1500, 1000]

plt.pie(revenue, labels=regions, autopct='%1.1f%%', colors=['red', 'orange', 'gold', 'coral'])
plt.title('Revenue Contribution by Region')
plt.legend()
plt.show()
```
