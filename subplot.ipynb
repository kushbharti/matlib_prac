```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4]
y = [10, 20, 15, 25]


plt.subplot(1,2,1) # row, column, 1st subplot
plt.plot(x,y)
plt.title('Line Chart')

plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('Bar Chart')

plt.tight_layout()
```

```python
fig, ax=plt.subplots(1, 2, figsize=(10, 5))

x = [1, 2, 3, 4]
y = [10, 20, 15, 25]

ax[0].plot(x,y, color='blue')
ax[0].set_title('Line Chart')

ax[1].bar(x, y, color='green')
ax[1].set_title('Bar Chart')

fig.suptitle('Comparison of Line and Bar Charts')

plt.tight_layout()

```
